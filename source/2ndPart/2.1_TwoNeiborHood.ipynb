```python
#matlab  groudTrue
matfn='gt_tianhui1200.mat';
data=sio.loadmat(matfn);
mk = data['gt']
print(mk)

moban=mk*0.0;

plt.figure(figsize=(6,6))
im = Image.fromarray(mk);
plt.imshow(im,cmap=plt.cm.get_cmap('cubehelix_r',7));
cbar=plt.colorbar()
cbar.set_ticks(np.linspace(0,7,8))
cbar.set_ticklabels(('  ','water','grass','bulding','road','luodi','background'))
plt.clim(-0.5,6.5)
plt.title('GroundTruth');
plt.show();
```

```python
Rows=Tpan.shape[0];  cols=Tpan.shape[1]; 
EarthGround=ABL.TwoPixcel();
NewMK=mk.reshape(1,mk.shape[0]*mk.shape[1]);

ALLUvalue=np.unique(EarthGround,return_index=True,return_inverse=True,return_counts = True)[0];
ALLUinverse=np.unique(EarthGround,return_index=True,return_inverse=True,return_counts = True)[2];

LenthofALLunique=len(ALLUvalue);
```

```python
ALLfinalF=[];  ALLTwoDimensionalFeatures=[];  ALLTwoD_Frequency=[];
for i in range(1,7):
    TF=(NewMK==i);
    TempMask=TF*EarthGround;
    usefulValue=TempMask[TF];

    Uvalue=np.unique(usefulValue,return_index=True,return_inverse=True,return_counts = True)[0];
    Uinverse=np.unique(usefulValue,return_index=True,return_inverse=True,return_counts = True)[2];
    Ufrence=(np.unique(usefulValue,return_index=True,return_inverse=True,return_counts = True)[3])/len(usefulValue);
    
    ALLTwoDimensionalFeatures.append(Uvalue); ALLTwoD_Frequency.append(Ufrence);

    fram1=pd.DataFrame({'id':ALLUvalue});
    fram2=pd.DataFrame({'id':Uvalue,'F':Ufrence});
   
    MEge=pd.merge(fram1,fram2,on='id',how='outer');
    MegeFILL=MEge.fillna({'F':0.0});
    finalF=MegeFILL['F'];
    
    ALLfinalF.append(finalF);
```

# 

```python
fig=plt.figure(figsize=(8,7));
ax=Axes3D(fig);
COR=['r','g','b','y','k','c'];  M=['^','o','8','p','d','*'];
for  LI in range(0,len(ALLTwoDimensionalFeatures)):
    ClassValue=ALLTwoDimensionalFeatures[LI];
    ClassFrequency=1000000*ALLTwoD_Frequency[LI];
    TempAllX=[]; TempAllY=[];
    for Ti in range(0,len(ClassValue)):
        TupleXY=ClassValue[Ti]; X=TupleXY[0]; Y=TupleXY[1];
        TempAllX.append(X);   TempAllY.append(Y);
    ax.scatter3D(TempAllX,TempAllY,ClassFrequency,c=COR[LI],marker=M[LI]);
ax.set_xlabel('B'); ax.set_ylabel('Bright'); ax.set_zlabel('Frequency');
ax.set_xlim(0,255); ax.set_ylim(0,255); ax.set_zlim(0,10000); ax.view_init(elev=20);
plt.show();
```

```python
LastLabel=np.zeros(LenthofALLunique);

for j in range(0,LenthofALLunique):
    MAXP=0;   maxindex=0;
    tempF=[ALLfinalF[0][j],ALLfinalF[1][j],ALLfinalF[2][j],ALLfinalF[3][j],ALLfinalF[4][j],ALLfinalF[5][j]];
    for k in range(0,len(tempF)):
        if(tempF[k]>MAXP):
            MAXP=tempF[k]
            maxindex=k+1;
    LastLabel[j]=maxindex;
```

```python
AllX=[];  AllY=[];
for Si in range(0,LenthofALLunique):
    Tuplexy=ALLUvalue[Si]; x=Tuplexy[0]; y=Tuplexy[1];
    AllX.append(x);    AllY.append(y);

plt.figure(figsize=(8,7));
plt.xlim(0,255); plt.ylim(0,255); 
plt.title('Two-dimensional features-classLabel');  plt.xlabel('B');  plt.ylabel('Bright');    
plt.scatter(x=AllX,y=AllY,c=LastLabel,marker='^',cmap=plt.cm.get_cmap('cubehelix_r',7));
cbar=plt.colorbar();
cbar.set_ticks(np.linspace(0,7,8));
cbar.set_ticklabels(('  ','water','grass','bulding','road','luodi','background'));
plt.show() 
```
